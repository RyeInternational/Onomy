```python
%matplotlib inline
#import packages
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import statistics
import seaborn as sns
df = pd.read_csv(r"C:\Users\jim13\Documents\project1\BusinessFinance.csv")
df.drop("Unnamed: 11",axis=1,inplace=True)
df.drop("Unnamed: 12",axis=1,inplace=True)
df.drop("isPaid",axis=1,inplace=True)
df.drop("length_for_calculation",axis=1,inplace=True)
df.drop("unit_for_calculation",axis=1,inplace=True)
df.drop("publishedTime",axis=1,inplace=True)
df.drop("url",axis=1,inplace=True)
df.drop("id",axis=1,inplace=True)
df.drop("contentInfo",axis=1,inplace=True)
df['price'].replace('Free','0',inplace=True)
df.head(999)
```
