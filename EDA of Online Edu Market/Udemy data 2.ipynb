```python
%matplotlib inline
#import packages
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import statistics
import seaborn as sns
df = pd.read_csv(r"C:\Users\jim13\Documents\project1\udemy.csv")
df.drop("id",axis=1,inplace=True)
df.drop("avg_rating_recent",axis=1,inplace=True)
df.drop("is_paid",axis=1,inplace=True)
df.drop("url",axis=1,inplace=True)
df.drop("avg_rating",axis=1,inplace=True)
df.drop("num_published_practice_tests",axis=1,inplace=True)
df.drop("created",axis=1,inplace=True)
df.drop("published_time",axis=1,inplace=True)
df.drop("discount_price__amount",axis=1,inplace=True)
df.drop("discount_price__currency",axis=1,inplace=True)
df.drop("discount_price__price_string",axis=1,inplace=True)
df.drop("price_detail__currency",axis=1,inplace=True)
df.drop("price_detail__price_string",axis=1,inplace=True)
df.drop("is_wishlisted",axis=1,inplace=True)

df.head(99)
```
